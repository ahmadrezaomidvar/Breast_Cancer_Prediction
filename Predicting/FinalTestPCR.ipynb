```python
df_original = pd.read_excel('./testDataset.xls')
df_original.head()
df = df_original.drop(labels=['ID'], axis=1, inplace=False)
for col in df.columns:
    df[col] = df[col].replace(999, np.nan)
    df[col] = df[col].fillna(df[col].mean())
df = df.astype(float, copy=False)
df.head()
```

```python
# load files
with open('./selected_columns.txt', 'r') as f:
    selected_columns = f.read().splitlines()

scaler = load('./sc_fs.pkl')
model = load('./svm_model.pkl')
```

```python
# Predict
df = df[selected_columns]
df_scaled = scaler.transform(df)

y_pred = model.predict(df_scaled)
prediction = df_original['ID']
prediction = pd.concat([prediction, pd.DataFrame(y_pred)], axis=1)
prediction.columns = ['ID', 'prediction']

prediction.to_csv('./PCR_prediction.csv', index=False)
```
